##count()

```python
#Returns the number of elements with the specified value
```

```python
numbers = [1, 2, 3, 4, 1, 2, 1, 3]
count_ones = numbers.count(1)
print(count_ones)
```

```python
fruits = ['apple', 'banana', 'orange', 'apple']
count_apples = fruits.count('apple')
print(count_apples)
```

```python
mixed_data = [1, 'apple', True, 3.14, 'apple', 1, 'world']
count_apple = mixed_data.count('apple')
print(count_apple)
```

```python
empty_list = []
count_empty = empty_list.count(42)
print(count_empty)
```

```python
names = ['Alice', 'Bob', 'Charlie', 'Alice', 'David', 'Alice']
count_alice = names.count('Alice')
print(count_alice)
```

```python
animals = ['dog', 'cat', 'dog', 'elephant', 'dog', 'lion']
count_dogs = animals.count('dog')
print(count_dogs)
```

```python
mixed_data = [1, 'apple', True, 3.14, 'apple', 1, 'world']
count_true = mixed_data.count(True)
print(count_true)
```

```python
original_list = [1, 2, 3, 4, 5, 2, 2, 1]
count_two = original_list.count(2)
print(count_two)
```

```python
names = ['Alice', 'Bob', 'Charlie', 'David', 'Alice', 'Charlie']
count_charlie = names.count('Charlie')
print(count_charlie)
```

```python
numbers = [5, 5, 5, 5, 5]
count_five = numbers.count(5)
print(count_five)
```

## extend()

```python
#Add the elements of a list (or any iterable), to the end of the current list
```

```python
fruits = ['apple', 'banana', 'orange']
more_fruits = ['kiwi', 'mango', 'pineapple']
fruits.extend(more_fruits)
print(fruits)
```

```python
numbers = [1, 2, 3]
more_numbers = [4, 5, 6]
numbers.extend(more_numbers)
print(numbers)
```

```python
letters = ['a', 'b', 'c']
vowels = ['e', 'i', 'o', 'u']
letters.extend(vowels)
print(letters)
```

```python
empty_list = []
numbers = [10, 20, 30]
empty_list.extend(numbers)
print(empty_list)
```

```python
words = ['hello', 'world']
new_words = 'python'
words.extend(new_words)
print(words)
```

```python
colors = ['red', 'green', 'blue']
additional_colors = ('yellow', 'purple')
colors.extend(additional_colors)
print(colors)
```

```python
languages = ['Python', 'Java']
languages.extend('C#')
print(languages)
```

```python
nested_list = [[1, 2], [3, 4]]
new_elements = [5, 6]
nested_list.extend(new_elements)
print(nested_list)
```

```python
mixed_data = [1, 'apple', True]
more_data = {'key': 'value'}
mixed_data.extend(more_data)
print(mixed_data)
```

```python
alphabets = ['a', 'b']
alphabets.extend('cd')
print(alphabets)
```

```python
names = ['Alice', 'Bob']
more_names = ('Charlie', 'David')
names.extend(more_names)
print(names)
```

# index()

```python
numbers = [10, 20, 30, 20, 40]
index_20 = numbers.index(20)
print(index_20)
```

```python
mixed_data = [1, 'apple', True, 'banana']
index_true = mixed_data.index(True)
print(index_true)
```

```python
alphabets = ['a', 'b', 'c', 'a']
index_a = alphabets.index('a')
print(index_a)
```

```python
nested_list = [[1, 2], [3, 4]]
index_3 = nested_list.index([3, 4])
print(index_3)
```

```python
colors = ['red', 'green', 'blue']
index_green = colors.index('green')
print(index_green)
```

```python
names = ['Alice', 'Bob', 'Alice']
index_alice = names.index('Alice')
print(index_alice)
```

```python
colors = ['red', 'green', 'blue', 'green']
index_green_2 = colors.index('green', 2)
print(index_green_2)
```

```python
alphabets = ['a', 'b', 'c']
index_b = alphabets.index('b')
print(index_b)
```

```python
numbers = [10, 20, 30, 20, 40]
index_20_after_2 = numbers.index(20, 2)
print(index_20_after_2)
```

## pop()
